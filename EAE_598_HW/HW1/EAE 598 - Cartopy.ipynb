### EAE 598 - Jupyter Notebook: Explaining Cartopy

#### As discussed in class, Cartopy is a library in Python that assists with geographic projections. It is developed on top of the popular data visualization library, Matplotlib, so they work nicely together. 

The first step is to import Cartopy and Matplotlib

```python
import cartopy.crs as ccrs 
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
```

#### First Example: Basic Plotting

#### Required Steps

#### 1) Define your projection (e.g., PlateCarree)
#### 2) Add cfeature for different features such as country and state borders as well as oceans, lakes, and rivers

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.add_feature(cfeature.BORDERS.with_scale('50m'), linewidth=0.5)
```

#### That plot looks a bit incomplete, particularly around the coastlines so let's add those!

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.add_feature(cfeature.BORDERS.with_scale('50m'), linewidth=0.5)
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
```

#### Much better, but let's focus on the United States

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.add_feature(cfeature.BORDERS.with_scale('50m'), linewidth=0.5)
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
ax.set_extent([-125, -67, 24, 49])
```

#### Good, but what about the states?

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.add_feature(cfeature.BORDERS.with_scale('50m'), linewidth=0.5)
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
ax.set_extent([-125, -67, 24, 49])
ax.add_feature(cfeature.STATES.with_scale('50m'), linewidth=0.5)
```

#### Let's add some color to the land and water!

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
ax.add_feature(cfeature.BORDERS.with_scale('50m'), linewidth=0.5)
ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
ax.set_extent([-125, -67, 24, 49])
ax.add_feature(cfeature.STATES.with_scale('50m'), linewidth=0.5)
ax.add_feature(cfeature.LAKES, zorder=1, color='#ecf9fd')
ax.add_feature(cfeature.OCEAN, color='#ecf9fd')
ax.add_feature(cfeature.LAND, color='#fbf5e9')
```

#### Example 3: ERA5 Data with Cartopy

```python
import xarray as xr
from siphon.catalog import TDSCatalog
```

#### Read in the ERA5 temperature data for 2/16/2021 using Xarray and select the temperature at 1000-hPa

```python
era5_url = 'https://thredds.rda.ucar.edu/thredds/catalog/files/g/d633000/e5.oper.an.pl/202102/catalog.html?dataset=files/g/d633000/e5.oper.an.pl/202102/e5.oper.an.pl.128_130_t.ll025sc.2021021600_2021021623.nc'
era5_ds = TDSCatalog(era5_url)
era5_ds = era5_ds.datasets[0].access_urls['OPENDAP']
ds = xr.open_dataset(era5_ds)
t_1000 = ds['T'].sel(level=1000) # units: K
```

#### Iterate over the time dimension and plot the temperature at 1000-hPa with the PlateCarree projection

```python
for i in range(0, 1):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
    ax.add_feature(cfeature.BORDERS.with_scale('50m'), linewidth=0.5)
    ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
    t_1000.isel(time=i).plot(ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm', cbar_kwargs={'label': 'Temperature (K)'})
    plt.title('ERA5 Temperature Analysis')
    plt.show()
```

Okay, but let's focus on the United States 

```python
for i in range(0, 1):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.PlateCarree())
    ax.add_feature(cfeature.BORDERS.with_scale('50m'), linewidth=0.5)
    ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
    ax.add_feature(cfeature.STATES.with_scale('50m'), linewidth=0.5)
    ax.set_extent([-125, -67, 24, 49])
    t_1000.isel(time=i).plot(ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm', cbar_kwargs={'label': 'Temperature (K)'})
    plt.title('ERA5 Temperature Analysis')
    plt.show()
```

Quite a rough projection... need to specify the central longitude and latitude to focus on North America

```python
for i in range(0, 1):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1, projection=ccrs.Orthographic(central_longitude=-95, central_latitude=37.5))
    ax.add_feature(cfeature.BORDERS.with_scale('50m'), linewidth=0.5)
    ax.add_feature(cfeature.COASTLINE.with_scale('50m'), linewidth=0.5)
    ax.add_feature(cfeature.STATES.with_scale('50m'), linewidth=0.5)
    t_1000.isel(time=i).plot(ax=ax, transform=ccrs.PlateCarree(), cmap='coolwarm', cbar_kwargs={'label': 'Temperature (K)'})
    plt.title('ERA5 Temperature Analysis')
    plt.show()
```
